**Вычисление показателя RSI (Relative Strength Index)**. RSI - это технический анализатор, который используется для оценки силы и направления цены актива. Для вычисления показателя RSI можно использовать следующий код:

```python
import pandas as pd
import yfinance as yf

def rsi(data, period=14):
    delta = data.diff()
    up = delta.where(delta > 0, 0)
    down = -delta.where(delta < 0, 0)
    gain = up.rolling(window=period).mean()
    loss = down.rolling(window=period).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

msft = yf.Ticker("PINS")
hist = msft.history(period="max")
rsi = rsi(hist['Close'])
print(rsi.tail())
```

**Вычисление показателя MACD (Moving Average Convergence Divergence)**. MACD - это технический индикатор, который используется для определения тренда и возможных точек входа и выхода на рынок. Для вычисления показателя MACD можно использовать следующий код:

```python
import pandas as pd
import yfinance as yf

def macd(data, fast=12, slow=26, signal=9):
    ema_fast = data.ewm(span=fast, adjust=False).mean()
    ema_slow = data.ewm(span=slow, adjust=False).mean()
    macd = ema_fast - ema_slow
    signal_line = macd.ewm(span=signal, adjust=False).mean()
    histogram = macd - signal_line
    return macd, signal_line, histogram

msft = yf.Ticker("PINS")
hist = msft.history(period="max")
macd, signal, histogram = macd(hist['Close'])
print(macd.tail())
print(signal.tail())
print(histogram.tail())
```

**Использование алгоритмов машинного обучения для предсказания будущих цен.** Для этого можно использовать библиотеки Python, такие как scikit-learn, tensorflow или pytorch. Например, для обучения модели линейной регрессии на исторических данных о биржевых котировках можно использовать следующий код:

```python
import pandas as pd
import yfinance as yf
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

msft = yf.Ticker("PINS")
hist = msft.history(period="max")

X = pd.DataFrame(index=hist.index)
X['Open'] = hist['Open']
X['High'] = hist['High']
X['Low'] = hist['Low']
y = hist['Close']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
```

**Получение исторических данных о котировках Pinterest с помощью библиотеки yfinance**. Затем мы создаем набор признаков, состоящий из цен открытия, максимальной и минимальной цен за каждый день. Целевой переменной является цена закрытия. Мы разделяем данные на обучающую и тестовую выборки с помощью функции train_test_split из библиотеки scikit-learn. Затем мы создаем модель линейной регрессии и обучаем ее на обучающих данных. Наконец, мы делаем прогноз на тестовых данных и вычисляем среднеквадратичную ошибку (MSE) для оценки качества модели.


```python
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error:", mse)
```

В следующем примере мы используем библиотеку yfinance для получения исторических данных о котировках Pinterest и Amazon. Затем мы создаем общий набор данных, содержащий цены закрытия обеих компаний, и кластеризуем его с помощью алгоритма KMeans. Наконец, мы отображаем результаты кластеризации на графике, используя библиотеку Matplotlib.

```python
import yfinance as yf
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Получение исторических данных о котировках Pinterest и Amazon
msft = yf.download("PINS", start="2020-01-01", end="2024-01-01")
aapl = yf.download("AMZN", start="2020-01-01", end="2024-01-01")

# Создание общего набора данных
data = pd.concat([msft['Close'], aapl['Close']], axis=1)
data.columns = ['PINS', 'AMZN']

# Кластеризация с помощью алгоритма KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

# Визуализация результатов кластеризации
plt.scatter(data['PINS'], data['AMZN'], c=kmeans.labels_, cmap='rainbow')
plt.xlabel('PINS')
plt.ylabel('AMZN')
plt.show()
```
